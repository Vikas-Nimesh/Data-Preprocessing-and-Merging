```python
import numpy as np
import pandas as pd
```

# Filtering, Removing and Filling in Missing Data 

```python
data = pd.DataFrame([[1,None,3,4,5],[6,None,8,9,10],[11,12,13,14,15]])
data
```

# Filtering out missing data 

```python
data.isnull()
```

```python
data.notnull()
```

# Removing Nan values

```python
data.dropna() #this will remove all the rows which includes Nan values
```

```python
data.dropna(axis=1) #this will remove all the columns which includes Nan values
```

###### Note:
dropna(how = 'all') this will remove all rows only if all rows have Nan value.<br>
dropna(thresh = 3) this will remove all rows only if all rows have =<3 value.

# Filling in missing data

```python
temp = pd.read_csv('data/temperature.csv',index_col= 'time')
temp
```

```python
temp.fillna(20) #this will fills 'Nan' values with '20'
#temp.fillna({'day1':20,'day2':30,'day4':40,}) #this will fills custom values according to passed dictonary
```

```python
temp.fillna(method='ffill') #this will propagate last valid observation forward
#temp.fillna(method='bfill') #it will backward fill the NaN values that are present.
```

```python
temp.fillna(temp.mean()) #this will fills values with mean of theri corresponding column
```
