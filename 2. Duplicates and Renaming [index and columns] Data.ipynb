# Duplicates and Renaming [index and columns] Data 

# Duplicates

## Finding duplicates

```python
data = pd.read_csv('data/ex1.csv')
data.head()
```

```python
data.duplicated()
```

Note: For large dataset it hard to notice duplicated values this above so we uses any() function which return true or false if any duplicate is present or not.

```python
data.duplicated().any()
```

```python
data[data.duplicated()]
```

## Removing duplicates

```python
data.drop_duplicates(inplace = True)
```

```python
data.reset_index(drop = True)
```
